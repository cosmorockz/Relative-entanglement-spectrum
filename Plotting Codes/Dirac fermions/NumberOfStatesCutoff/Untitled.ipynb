```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mz1 = 0
mz2 = 2
maxN = 30
windowL = 0.40
windowU = 0.60
```

```python
CC1, CC2 = [], []
NN1, NN2 = [], []

for i in range(1,maxN+1):
    M,E = np.loadtxt("ES_DF_m_"+str(mz1)+"_cutoff_"+str(i)+".dat",unpack=True)
    n = 0
    for j in range(len(E)):
        if E[j]>=windowL and E[j]<=windowU:
            n += 1
    CC1.append(i)
    NN1.append(n)

for i in range(1,maxN+1):
    M,E = np.loadtxt("ES_DF_m_"+str(mz2)+"_cutoff_"+str(i)+".dat",unpack=True)
    n = 0
    for j in range(len(E)):
        if E[j]>=windowL and E[j]<=windowU:
            n += 1
    CC2.append(i)
    NN2.append(n)
```

```python
plt.scatter(CC1,NN1)
plt.scatter(CC2,NN2)
```

```python
M1, E1 = np.loadtxt("ES_DF_m_"+str(mz1)+"_cutoff_"+str(10)+".dat",unpack=True)
M2, E2 = np.loadtxt("ES_DF_m_"+str(mz2)+"_cutoff_"+str(10)+".dat",unpack=True)
```

```python
import matplotlib.gridspec as gridspec
from matplotlib.patches import Rectangle

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.rc('axes', axisbelow=True)

# fig,ax1 = plt.subplots(1,1,sharex=True,sharey=True)
fig = plt.figure(figsize=(10,6),tight_layout=True)
gs = gridspec.GridSpec(2, 2)
ax1 = fig.add_subplot(gs[:, 1])
ax21 = fig.add_subplot(gs[0, 0])
ax22 = fig.add_subplot(gs[1, 0])
# ax1 = fig.gca()

'''Plot 1'''
ax1.grid(which='both')
# Major ticks every 20, minor ticks every 5
xmajor_ticks = np.arange(0, 45, 15)
xminor_ticks = np.arange(0, 33.75, 3.75)
ymajor_ticks = np.arange(0, 45, 15)
yminor_ticks = np.arange(0, 33.75, 3.75)

ax1.set_xticks(xmajor_ticks)
ax1.set_xticks(xminor_ticks, minor=True)
ax1.set_yticks(ymajor_ticks)
ax1.set_yticks(yminor_ticks, minor=True)

ax1.yaxis.set_tick_params(width=2, length=5)
ax1.xaxis.set_tick_params(width=2, length=5)
for axes in ['top', 'bottom', 'left', 'right']:
    ax1.spines[axes].set_linewidth(0.5)
    
ax1.scatter(CC1,NN1,s=24,label="$M=$"+str(mz1),color="#2E4272")
ax1.scatter(CC2,NN2,s=24,label="$M=$"+str(mz2),color="green")
ax1.legend(loc="upper left")
ax1.set_xlabel("Cutoff",color="#061539")
ax1.set_ylabel("No. of States",color="#061539")
ax1.text(-6,32,"(c)",fontsize=20,fontweight="bold",va="top",ha="right")

'''Plot 2'''
ax21.grid(which='both')
# Major ticks every 20, minor ticks every 5
xmajor_ticks = np.arange(-10, 20, 10)
xminor_ticks = np.arange(-10, 12.5, 2.5)
ymajor_ticks = np.arange(0, 1.5, 0.5)
yminor_ticks = np.arange(0, 1.25, 0.25)

ax21.set_xticks(xmajor_ticks)
ax21.set_xticks(xminor_ticks, minor=True)
ax21.set_yticks(ymajor_ticks)
ax21.set_yticks(yminor_ticks, minor=True)

ax21.yaxis.set_tick_params(width=2, length=5)
ax21.xaxis.set_tick_params(width=2, length=5)
for axes in ['top', 'bottom', 'left', 'right']:
    ax21.spines[axes].set_linewidth(0.5)
    
ax21.scatter(M1,E1,s=24,color="#2E4272")
ax21.set_xlabel("$\ell$",color="#061539")
ax21.set_ylabel("$\lambda$",color="#061539")
ax21.text(-14,1.1,"(a)",fontsize=20,fontweight="bold",va="top",ha="right")

ax21.add_patch(Rectangle((-10,windowL),height=windowU-windowL,width=20,color='grey',alpha=0.4))

ax21.set_xlim(-10,10)

'''Plot 3'''
ax22.grid(which='both')
# Major ticks every 20, minor ticks every 5
xmajor_ticks = np.arange(-10, 20, 10)
xminor_ticks = np.arange(-10, 12.5, 2.5)
ymajor_ticks = np.arange(0, 1.5, 0.5)
yminor_ticks = np.arange(0, 1.25, 0.25)

ax22.set_xticks(xmajor_ticks)
ax22.set_xticks(xminor_ticks, minor=True)
ax22.set_yticks(ymajor_ticks)
ax22.set_yticks(yminor_ticks, minor=True)

ax22.yaxis.set_tick_params(width=2, length=5)
ax22.xaxis.set_tick_params(width=2, length=5)
for axes in ['top', 'bottom', 'left', 'right']:
    ax22.spines[axes].set_linewidth(0.5)
    
ax22.scatter(M2,E2,s=24,color="green")
ax22.set_xlabel("$\ell$",color="#061539")
ax22.set_ylabel("$\lambda$",color="#061539")
ax22.text(-14,1.1,"(b)",fontsize=20,fontweight="bold",va="top",ha="right")

ax22.add_patch(Rectangle((-10,windowL),height=windowU-windowL,width=20,color='grey',alpha=0.4))

ax22.set_xlim(-10,10)





# plt.xlabel("m",color="#061539")
# plt.xlim(-10,10)
plt.savefig("ESCutoffDependence.pdf",bbox_inches="tight",pad_inches=0.4,dpi=300)
plt.show()
```
