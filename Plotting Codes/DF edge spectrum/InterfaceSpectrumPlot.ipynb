```python
import numpy as np
import mpmath as mp
```

```python
import matplotlib.pyplot as plt
```

```python
M = [2,4]
Delta = [0,0]
MM, E = np.loadtxt("MA_"+str(M[0])+",MB_"+str(M[1])+",DeltaA_"+str(Delta[0])+",DeltaB_"+str(Delta[1])+".dat",usecols=(0,1),unpack=True)
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 20
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.rc('axes', axisbelow=True)

im = plt.imread("case1.PNG")

fig,ax1 = plt.subplots(1,1,sharex=True,sharey=True)
ax = fig.gca()
ax1.grid(which='both')
# Major ticks every 20, minor ticks every 5
xmajor_ticks = np.arange(-10, 15, 10)
xminor_ticks = np.arange(-10, 15, 2.5)
ymajor_ticks = np.arange(-10, 15, 10)
yminor_ticks = np.arange(-10, 15, 2.5)

ax.set_xticks(xmajor_ticks)
ax.set_xticks(xminor_ticks, minor=True)
ax.set_yticks(ymajor_ticks)
ax.set_yticks(yminor_ticks, minor=True)

ax1.yaxis.set_tick_params(width=2, length=5)
ax1.xaxis.set_tick_params(width=2, length=5)
for axes in ['top', 'bottom', 'left', 'right']:
    ax1.spines[axes].set_linewidth(0.5)

left, bottom, width, height = [0.1, 0.36, 0.3, 0.3]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.imshow(im)
ax2.axis("off")
# ax2.scatter(MM,EE,s=1,color="green")
# ax2.set_xticks([])
# ax2.set_yticks([])
# ax2.set_xlabel("$m$")
# ax2.set_ylabel("$E$")

ax1.scatter(MM,E,s=24,color="#2E4272")
ax1.set_xlabel("$m$",color="#061539")
ax1.set_ylabel("$E$",color="#061539")
# plt.xlabel("m",color="#061539")
plt.savefig("./MA"+str(M[0])+"MB"+str(M[1])+"DeltaA"+str(Delta[0])+"DeltaB"+str(Delta[1])+".pdf",bbox_inches="tight",pad_inches=0.4,dpi=300)
plt.show()
```
