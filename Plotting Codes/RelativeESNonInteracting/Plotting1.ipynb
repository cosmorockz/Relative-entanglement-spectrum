```python
import numpy as np
import mpmath as mp
```

```python
import matplotlib.pyplot as plt
```

```python
MM, E = np.loadtxt("Mass4Delta-4Cutoff10.dat",usecols=(0,1),unpack=True)
```

```python
SMALL_SIZE = 30
MEDIUM_SIZE = 30
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.rc('axes', axisbelow=True)

im = plt.imread("cc_new.PNG")

fig,ax1 = plt.subplots(1,1,sharex=True,sharey=True)
ax1 = fig.gca()
ax1.grid(which='both')
# Major ticks every 20, minor ticks every 5
xmajor_ticks = np.arange(-10, 15, 10)
xminor_ticks = np.arange(-10, 15, 2.5)
ymajor_ticks = np.arange(0, 1.5, 0.5)
yminor_ticks = np.arange(0, 1.5, 0.25)

ax1.set_xticks(xmajor_ticks)
ax1.set_xticks(xminor_ticks, minor=True)
ax1.set_yticks(ymajor_ticks)
ax1.set_yticks(yminor_ticks, minor=True)

ax1.yaxis.set_tick_params(width=2, length=5)
ax1.xaxis.set_tick_params(width=2, length=5)
for axes in ['top', 'bottom', 'left', 'right']:
    ax1.spines[axes].set_linewidth(0.5)

left, bottom, width, height = [0.2, 0.76, 0.6, 0.6]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.imshow(im)
ax2.axis("off")
# ax2.scatter(MM,EE,s=1,color="green")
# ax2.set_xticks([])
# ax2.set_yticks([])
# ax2.set_xlabel("$m$")
# ax2.set_ylabel("$E$")

ax1.scatter(MM,E,s=24,color="#2E4272")
ax1.set_xlabel("$m$",color="#061539")
ax1.set_ylabel("$\lambda$",color="#061539")
# plt.xlabel("m",color="#061539")
plt.savefig("Mass4Delta-4.pdf",bbox_inches="tight",pad_inches=0.4,dpi=300)
plt.show()
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 20
BIGGER_SIZE = 20

plt.figure(1)

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.rc('axes', axisbelow=True)

im = plt.imread("aa_new.PNG")

fig,ax1 = plt.subplots(1,1,sharex=True,sharey=True)
ax = fig.gca()
ax1.grid(which='both')
# Major ticks every 20, minor ticks every 5
xmajor_ticks = np.arange(-10, 15, 10)
xminor_ticks = np.arange(-10, 15, 2.5)
ymajor_ticks = np.arange(0, 1.5, 0.5)
yminor_ticks = np.arange(0, 1.5, 0.25)

ax.set_xticks(xmajor_ticks)
ax.set_xticks(xminor_ticks, minor=True)
ax.set_yticks(ymajor_ticks)
ax.set_yticks(yminor_ticks, minor=True)

ax1.yaxis.set_tick_params(width=2, length=5)
ax1.xaxis.set_tick_params(width=2, length=5)
for axes in ['top', 'bottom', 'left', 'right']:
    ax1.spines[axes].set_linewidth(0.5)

left, bottom, width, height = [0.2, 0.76, 0.6, 0.6]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.imshow(im)
ax2.axis("off")
# ax2.scatter(MM,EE,s=1,color="green")
# ax2.set_xticks([])
# ax2.set_yticks([])
# ax2.set_xlabel("$m$")
# ax2.set_ylabel("$E$")

ax1.scatter(MM,E,s=24,color="#2E4272")
ax1.set_xlabel("$m$",color="#061539")
ax1.set_ylabel("$\lambda$",color="#061539")

x1, x2, y1, y2 = -2.5,2.5,0.25,0.75
axins = ax1.inset_axes([0.6, 0.36, 0.38, 0.5],frame_on=True)
axins.grid()
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.set_xticklabels([])
axins.set_yticklabels([])
axins.scatter(MM1,E1)

ax.indicate_inset_zoom(axins, edgecolor="black")
axins.text(0.2,0.64,"$\delta_S = 4$",size=20,color="red")

# plt.xlabel("m",color="#061539")
plt.savefig("M1_4M2_-4_Delpair4.pdf",bbox_inches="tight",pad_inches=0.4,dpi=300)
plt.show()
```
